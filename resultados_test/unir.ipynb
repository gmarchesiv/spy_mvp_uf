```python
import pandas as pd
import glob
import numpy as np

# Ruta donde est√°n tus archivos Excel (ajusta la ruta si es necesario)
ruta_archivos = "*.xlsx"

# Lista todos los archivos .xlsx
archivos = glob.glob(ruta_archivos)

# Crea una lista para almacenar los DataFrames
lista_df = []

# Lee cada archivo y agrega la hoja 'Resultados'
for archivo in archivos:
    df = pd.read_excel(archivo, sheet_name='Resultados')
    lista_df.append(df)

# Une todos los DataFrames en uno solo
df_final = pd.concat(lista_df, ignore_index=True)
# df_final.to_csv("resultados/resultado.csv",index=True)
```

```python
writer = pd.ExcelWriter(
    f"resultados/SPY_resultado.xlsx", engine="xlsxwriter"
)  # generamos el Archivos
```

```python

df_final.to_excel(writer, sheet_name="Resumen", index=True)
```

```python
writer.close()
```
